```python
sair =  'SIM'
candidatos = {
    'nomes': [],
    'idades':[],
    'cidades': [],
    'vagas':[]
}

while sair == 'SIM':
    nome = input('Nome: ')
    idade = int(input('idade: '))
    cidade = input('cidade')
    vaga = input('Vaga: ')

    candidatos['nomes'].append(nome)
    candidatos['idades'].append(idade)
    candidatos['cidades'].append(cidade)
    candidatos['vagas'].append(vaga)
    sair = input('Deseja cadastrar mais candidatos?').strip().upper()

for i in range(len(candidatos['nomes'])):
    print(f'O candidato {candidatos['nomes'][i]} tem {candidatos['idades'][i]}anos, mora na cidade de {candidatos["cidades"][i]} candidato a vaga {candidatos['vagas'][i]}')
```

```python
while True:
    try:
        a = float(input("Primeiro numero:"))
        b = float(input('Segundo numero: '))
        resultado = a / b
             
    except ValueError:
        print("Digite apenas numeros validos!")
    except ZeroDivisionError:
        print("Nao e possivel dividir por zero!")

    else:
        print(f"Resultado: {resultado}")
        break
```

```python
while True:
    try:
        a = float(input("Primeiro numero:"))
        operacao = input('Digite a operação: (+, -, *, /)')
        b = float(input('Segundo numero: '))
    
        if operacao not in ("+", "-", "*", "/"):
            print("Digite apenas numeros validos!")
            continue
    except ValueError:
          print("Erro: digite um número válido.")
          continue
    else:
        try:
            if operacao == "+":
                resultado = a + b
            elif operacao == "-":
                resultado = a - b
            elif operacao == "*":
                resultado = a * b
            elif operacao == "/":
                resultado = a / b

            print(f"Resultado: {resultado}")
            break

        except ZeroDivisionError:
            print("Erro: não é possível dividir por zero.")
```
