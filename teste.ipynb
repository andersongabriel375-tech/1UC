```python
temperaturas = []

for i in range(6):
    temp = float(input(f'Digite a primeira temperatura {i+1}'))
    temperaturas.append(temp)
for temp in temperaturas:
    if temp < 15:
        print(f'{temp}ºC frio')
    elif temp <= 28:
        print(f'{temp}°C - Agradável')
    else:
        print(f'{temp}ºC - Quente')
```
